```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_scores.csv")
```

## **Drop unnamed column**


```python
df = df.drop("Unnamed: 0", axis=1)
print(df.head())
```

## **Change weekly study hours column**

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct", "5-10")
df.head()
```

```python
df['NrSiblings'].plot(kind='hist', bins=20, title='NrSiblings')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

## **Gender distribution**

```python
plt.figure(figsize=(4,4))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender distribution")
plt.show()
```

**The no. of females in the data is bigger than the no. of males.**

```python
gb = df.groupby("ParentEduc").agg({"MathScore": 'mean', "ReadingScore": 'mean', "WritingScore": 'mean'})
print(gb)
```

```python
 plt.figure(figsize=(5,5))
 sns.heatmap(gb, annot=True) # cmap="YlGnBu"
 plt.title("Relationship between Parent's education and Student's score")
 plt.show()
```

**From the above chart it can be seen that the education of the parents it's reflected in the student's scores.**

```python
gb2 = df.groupby("ParentMaritalStatus").agg({"MathScore": 'mean', "ReadingScore": 'mean', "WritingScore": 'mean'})
print(gb2)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb2, annot=True, cmap="YlGnBu")
plt.title("Relationship between Parent's marital s and Student's score")
plt.show()
```

**From the above chart it can be seen that the marital status of the parents has no impact or negligable impact on the student score.**

```python
gb3 = df.groupby("WklyStudyHours").agg({"MathScore": 'mean', "ReadingScore": 'mean', "WritingScore": 'mean'})
print(gb3)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb3, annot=True, cmap="YlGnBu")
plt.title("Relationship between Weekly study hours and Student's score")
plt.show()
```

# Distribution of ethnic groups


```python
groupA = df.loc[(df['EthnicGroup'] == "group A" )].count()
groupB = df.loc[(df['EthnicGroup'] == "group B" )].count()
groupC = df.loc[(df['EthnicGroup'] == "group C" )].count()
groupD = df.loc[(df['EthnicGroup'] == "group D" )].count()
groupE = df.loc[(df['EthnicGroup'] == "group E" )].count()

l = ["group A", "group B", "group C", "group D", "group E" ]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
print(mlist)


plt.pie(mlist, labels = l, autopct='%1.2f%%')
plt.title('Distribution of ethnic groups')
plt.show()

# print(groupA)
# print(groupA["EthnicGroup"])  # = 2219


# a new dataframe is created, with values that come under group A
```
