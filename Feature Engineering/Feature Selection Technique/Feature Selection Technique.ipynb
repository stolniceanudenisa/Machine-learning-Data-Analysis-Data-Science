### What is Feature Selection Techniques
Feature Selection is one of the core concepts in machine learning which hugely impacts the performance of your model. The data features that you use to train your machine learning models have a huge influence on the performance you can achieve.
Irrelevant or partially relevant features can negatively impact model performance.

### Benefits of performing feature selection before modeling your data?

<b>Reduces Overfitting:</b> Less redundant data means less opportunity to make decisions based on noise.

<b>Improves Accuracy:</b> Less misleading data means modeling accuracy improves.

<b>Reduces Training Time:</b> fewer data points reduce algorithm complexity and algorithms train faster.

### Import Dataset

```python
df = pd.read_csv(r'F:\Feature Engineering\Univariate Selection\cardio_train.csv',sep = ";")
df
```

```python
df.shape
```

```python
df['yr'] = (df['age']/365).round(0)
```

```python
df['yr']
```

```python
df.describe()
```

```python
df.corr()
```

```python
df = df.drop(['yr'],axis = 1)
```

```python
x = df.iloc[:,:-1]
```

```python
y = df.iloc[:,12]
```

```python
y
```

```python
from sklearn.feature_selection import SelectKBest
```

```python
from sklearn.feature_selection import f_classif
```

```python
FIT_FEATURES = SelectKBest(score_func = f_classif)
```

```python
FIT_FEATURES.fit(x,y)
```

```python
pd.DataFrame(FIT_FEATURES.scores_)
```

```python
SCORE_COL = pd.DataFrame(FIT_FEATURES.scores_,columns = ['score value'])
```

```python
SCORE_COL
```

```python
NAME_COL = pd.DataFrame(x.columns)
```

```python
pd.DataFrame(NAME_COL)
```

```python
topr_features = pd.concat([NAME_COL,SCORE_COL], axis =1)
```

```python
topr_features
```

```python
topr_features.nlargest(8,'score value')
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=.3,random_state = 1)
```

```python
xtrain
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(xtrain,ytrain)
```

```python
rfc.score(xtest,ytest)
```

### Feature Importance

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
model = ExtraTreesClassifier()
```

```python
model.fit(x,y)
```

```python
model.feature_importances_
```

```python
top = pd.Series(model.feature_importances_,index = x.columns)
```

```python
top
```

```python
top.nlargest(10).plot(kind = 'barh')
```
